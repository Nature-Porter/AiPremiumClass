```python
import csv
import pkuseg

seg = pkuseg.pkuseg()

ds_comments = []

with open('/mnt/data_1/zfy/4/week7/homework/homework3/DMSC.csv', 'r') as file:
    reader = csv.DictReader(file)
    for idx, row in enumerate(reader):
        if idx >= 30000:
            break
        star = int(row['Star'])
        words = seg.cut(row['Comment']) 
        if star in range(1, 3):
            ds_comments.append((words, 0))
        elif star in range(4, 6):
            ds_comments.append((words, 1))

```

```python
import matplotlib.pyplot as plt

comments_len = [len(c) for c,v in ds_comments]
plt.hist(comments_len, bins=100)
plt.show()
```

```python
plt.boxplot(comments_len)
plt.show
```

```python

ds_comments = [c for c in ds_comments if len(c[0]) in range(10, 120)]
print(len(ds_comments))
```
