```python
import jieba
import csv
```

```python
ds_comments = []
# 读取csv文件
with open('DMSC.csv','r') as f:
    reader = csv.DictReader(f)
    for row in reader :
        star = int(row['Star'])
        if star in [1, 2, 4, 5]:
            words = jieba.lcut(row['Comment'])
            ds_comments.append((words, 1 if star > 2 else 0))

print(len(ds_comments))
    
```

分析评论长度，考虑截取过长或者过短的评论

```python
import matplotlib.pyplot as plt 

comment_len = [len(c) for c,v in ds_comments]
plt.hist(comment_len, bins=100)
plt.show()
```

```python
#评论长度在10-150之间
ds_comments = [c for c in ds_comments if len(c[0]) in range(10, 150)]
print(len(ds_comments))
```

```python
#保存成pkl文件
import pickle
with open('ds_comments.pkl', 'wb') as f:
    pickle.dump(ds_comments, f)
```
