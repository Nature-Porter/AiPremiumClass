###### 使用graphrag构建一篇小说（自主选择文档）的RAG知识图，实现本地和全局问题的问答。（截图代码运行结果）


1. 安装 graphrag ：pip install graphrag


2. 创建input目录用来存放文本，选择了一段头部科技公司的描述文章
mkdir -p ./graphrag_test/input


3. 初始化graphrag设置工作区： graphrag init --root graphrag_test
<br>graphrag_test目录中会新增 .env , settings.yaml,及prompts目录 文件
<br>.env : 设置api_key 与 base_url
<br>settings.yaml : 配置模型参数
<br>prompts目录： 存放自定义prompt模板
<br>使用国内模型，chat-model用deepseek，embedding-model用bge-m3,settings.yaml核心配置参数如下：

```python
default_chat_model:
    type: openai_chat # or azure_openai_chat
    api_base: https://api.deepseek.com
    auth_type: api_key # or azure_managed_identity
    api_key: sk-********* # set this in the generated .env file
    model: deepseek-chat
    encoding_model: cl100k_base # automatically set by tiktoken if left undefined
  default_embedding_model:
    type: openai_embedding # or azure_openai_embedding
    api_base: http://localhost:11434/v1
    auth_type: api_key # or azure_managed_identity
    api_key: sk-cc333445644
    model: bge-m3
    encoding_model: cl100k_base # automatically set by tiktoken if left undefined
```

```python
# 测试Ollama bge-m3模型
from openai import OpenAI

client = OpenAI(
    base_url='http://localhost:11434/v1/',      
    api_key='ollama',  # 这里随便写，但是api_key字段一定要有
)

result = client.embeddings.create(
    model='bge-m3:latest',  # 这里要修改成 你 ollama 启动模型的名称
    input='你好，世界',
)

print(result)
```

4. 创建索引：graphrag index --root graphrag_test
