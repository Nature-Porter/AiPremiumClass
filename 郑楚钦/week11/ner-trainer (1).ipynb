```python
from transformers import AutoModelForTokenClassification, AutoTokenizer,DataCollatorForTokenClassification,TrainingArguments, Trainer
from datasets import load_dataset
import torch

MODEL_NAME = 'google-bert/bert-base-chinese'
# DATASET_NAME = 'nlhappy/CLUE-NER'
DATASET_NAME = 'doushabao4766/msra_ner_k_V3'

def get_model_and_tokenizer(num_labels, model_name = MODEL_NAME):
    model = AutoModelForTokenClassification.from_pretrained(model_name, num_labels=num_labels)
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    return model,tokenizer

model,tokenizer = get_model_and_tokenizer(7)
out = tokenizer('床前明月光', return_tensors='pt')
for i in out:
    print(i,'=>', out[i])
print()
out = model(**out, labels = torch.tensor([0,1,0,2,5,4,5]))
for i in out:
    print(i,'=>', out[i])
print()

ds = load_dataset(DATASET_NAME)
ds['train'] = ds['train'].shard(num_shards=4, index=2)
ds['test'] = ds['test'].shard(num_shards=4, index=2)
print(ds)

for row in ds['test']:
    for i in row:
        print(i,'=>', row[i])
    # out = tokenizer([''.join(row['tokens'])], return_tensors='pt')
    # print(out)
    # out = model(**out)
    # print(out)
    break
```

```python

def mapper(item):
    # if len(item['tokens']) == 0:
        # print(item)
        # return
    input = tokenizer([''.join(it) for it in item['tokens']],truncation=True,padding=True,max_length=512,add_special_tokens=False,return_tensors="pt")
    # print(input)
    len_limit = input["input_ids"].shape[1]
    input['labels'] = [arr[:len_limit] for arr in item['ner_tags']]
    return input

ds2 = ds.map(mapper, batched=True)

ds2.set_format('torch', columns=['input_ids', 'token_type_ids', 'attention_mask', 'labels'])
print(ds2)

for row in ds2['test']:
    for i in row:
        print(i,'=>', row[i])
    break
```

```python
args = TrainingArguments(
    output_dir="ner_train",  # 模型训练工作目录（tensorboard，临时模型存盘文件，日志）
    num_train_epochs = 3,    # 训练 epoch
    save_safetensors=False,  # 设置False保存文件可以通过torch.load加载
    per_device_train_batch_size=32,  # 训练批次
    per_device_eval_batch_size=32,
    report_to='tensorboard',  # 训练输出记录
    eval_strategy="epoch",
)

data_collator = DataCollatorForTokenClassification(tokenizer=tokenizer,padding=True)

import evaluate

entites = ['O', 'PER', 'ORG', 'LOC']
tags = ['O']
for entity in entites[1:]:
    tags.append('B-' + entity.upper())
    tags.append('I-' + entity.upper())
print(tags)
# metric 方法
def compute_metrics(result):
    # result 是一个tuple (predicts, labels)
    # 获取评估对象
    seqeval = evaluate.load('seqeval')
    predicts,labels = result
    predicts = np.argmax(predicts, axis=2)
    # 准备评估数据
    predicts = [[tags[p] for p,l in zip(ps,ls) if l != -100]
                 for ps,ls in zip(predicts,labels)]
    labels = [[tags[l] for p,l in zip(ps,ls) if l != -100]
                 for ps,ls in zip(predicts,labels)]
    results = seqeval.compute(predictions=predicts, references=labels)
    return results


trainer = Trainer(model, args, train_dataset=ds2['train'], eval_dataset=ds2['test'],data_collator=data_collator,compute_metrics=compute_metrics)

trainer.train()
```

```python

def predict(text):
    out = tokenizer(text, return_tensors='pt',add_special_tokens=False).to('cuda')
    word_ids = out.word_ids(0)
    # print(word_ids)
    out = model(**out)
    predicts = np.argmax(out['logits'].cpu().detach().numpy(), axis=2)[0]
    print(predicts)
    tmp_tag = None
    tmp = []
    result = []
    for i in range(len(predicts)):
        predict_label = predicts[i]
        if predict_label - 1 == tmp_tag:
            tmp.append(text[word_ids[i]])
        else:
            if tmp_tag:
                result.append({'entity': tags[tmp_tag][2:], 'content': ''.join(tmp)})
                # print(tags[tmp_tag][2:], tmp)
            tmp_tag = None
            tmp = []
            if predict_label % 2:
                tmp_tag = predict_label
                tmp.append(text[word_ids[i]])
            elif predict_label > 0:
                tmp_tag = predict_label - 1
                tmp.append(text[word_ids[i]])
    print(result)

predict('澳大利亚和日本进行了友好交流。')
predict('全佛山谁最能打？')
predict('双方确定了今后发展中美关系的指导方针。')
```
